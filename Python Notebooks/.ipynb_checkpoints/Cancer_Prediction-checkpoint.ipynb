# Breast Cancer Classification Notebook

## Dataset Source
For this notebook, we use the **Breast Cancer Wisconsin (Diagnostic) dataset**.  
It is available directly from `sklearn.datasets` or can be downloaded from the [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic)).


### Importing Libraries
We start by importing the required Python libraries for data analysis, visualization, and model building.


```python
# pip install ucimlrepo
```

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve
)
import pickle
from scipy.stats import randint

sns.set_style("whitegrid")
plt.rcParams['figure.dpi'] = 120

RANDOM_STATE = 42
```

### Loading Dataset

```python
# We fetch the dataset programmatically so the notebook is reproducible.
# The fetch_ucirepo call returns dataframes for features and targets.
from ucimlrepo import fetch_ucirepo

print("Fetching Breast Cancer Wisconsin (Diagnostic) dataset from UCI repository...")
breast = fetch_ucirepo(id=17)   # id=17 as you provided
X = breast.data.features        # pandas DataFrame of features
y = breast.data.targets         # pandas Series or DataFrame of targets (0/1)
meta = breast.metadata

print("\n--- Metadata ---")
print(meta)
print("\nFeature shape:", X.shape)
print("Target shape:", np.shape(y))

# Convert y to a Series if it's DataFrame-like
if not isinstance(y, pd.Series):
    y = pd.Series(np.ravel(y), name='target')

# Combine for convenience in EDA
df = pd.concat([X.reset_index(drop=True), y.reset_index(drop=True)], axis=1)

# Show first rows
print("\nFirst 5 rows of the combined DataFrame:")
display(df.head())
```

```python
# Print dataframe info & descriptive stats, show class imbalance,
# and draw colorful visualizations to match the style you requested.
print("\nDataFrame Info:")
df.info()

print("\nDescriptive Statistics (numerical columns):")
display(df.describe().T)

print("\nClass distribution (target counts):")
print(df['target'].value_counts())   # 0 = benign, 1 = malignant (per UCI/scikit convention)

# Colorful class distribution plot
plt.figure(figsize=(6,4))
sns.countplot(x='target', data=df, hue='target', palette=['#2ecc71', '#e67e22'], legend=False)
plt.xticks([0,1], ['Benign (0)', 'Malignant (1)'])
plt.title("Class Distribution — Benign vs Malignant")
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.show()
```

### Exploratory Data Analysis (EDA)
We perform basic checks and exploratory steps to understand the dataset better.

```python
# Print dataframe info & descriptive stats, show class imbalance,
# and draw colorful visualizations to match the style you requested.
print("\nDataFrame Info:")
df.info()

print("\nDescriptive Statistics (numerical columns):")
display(df.describe().T)

print("\nClass distribution (target counts):")
print(df['target'].value_counts())   # 0 = benign, 1 = malignant (per UCI/scikit convention)

# Colorful class distribution plot
plt.figure(figsize=(6,4))
sns.countplot(x='target', data=df, hue='target', palette=['#2ecc71', '#e67e22'], legend=False)
plt.xticks([0,1], ['Benign (0)', 'Malignant (1)'])
plt.title("Class Distribution — Benign vs Malignant")
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.show()
```

```python
# Combine features and target into one DataFrame
df = X.copy()
df['target'] = y

# Ensure target is numeric
df['target'] = df['target'].map({'B': 0, 'M': 1})

# Compute correlations only on numeric data
numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(16,14))
corr = numeric_df.corr()
sns.heatmap(corr, annot=False, cmap='Spectral', linewidths=0.3)
plt.title("Feature Correlation Heatmap (Spectral)")
plt.show()

# Show top correlated features with target
print("\nTop features correlated with target (absolute):")
corr_with_target = corr['target'].abs().sort_values(ascending=False)
display(corr_with_target.head(15))
```

### Splitting Dataset
We split the data into training and testing sets to evaluate our model performance.


```python
# Split into training and test sets (80/20). We will scale features for Logistic Regression,
# but RandomForest will be run on raw features (scaling doesn't harm it if you prefer unified pipeline).
X_raw = df.drop(columns=['target'])
y = df['target']

X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X_raw, y, test_size=0.20, random_state=RANDOM_STATE, stratify=y
)

print("Train set shape:", X_train_raw.shape)
print("Test set shape:", X_test_raw.shape)
```

### Prepare scaled datasets for logistic regression baseline

```python
# Logistic Regression benefits from scaling. We'll fit a StandardScaler on training data
# and transform both train and test splits. We'll save the scaler only if you want to persist it.
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train_raw), columns=X_train_raw.columns, index=X_train_raw.index)
X_test = pd.DataFrame(scaler.transform(X_test_raw), columns=X_test_raw.columns, index=X_test_raw.index)
```

### Baseline: Logistic Regression

```python
# Train a Logistic Regression baseline so we have a simple model to compare with RandomForest.
log_model = LogisticRegression(max_iter=2000, random_state=RANDOM_STATE)
log_model.fit(X_train, y_train)

y_pred_log = log_model.predict(X_test)
log_acc = accuracy_score(y_test, y_pred_log)

print(f"Logistic Regression Accuracy: {log_acc*100:.2f}%")
print("\nClassification Report (Logistic Regression):")
print(classification_report(y_test, y_pred_log, digits=4))
```

```python
# Confusion matrix for logistic regression
cm_log = confusion_matrix(y_test, y_pred_log)
plt.figure(figsize=(6,5))
sns.heatmap(cm_log, annot=True, fmt='d', cmap='coolwarm', linewidths=1.5, linecolor='black')
plt.title("Logistic Regression — Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# ROC AUC for logistic
y_proba_log = log_model.predict_proba(X_test)[:,1]
roc_auc_log = roc_auc_score(y_test, y_proba_log)
print(f"Logistic Regression ROC AUC: {roc_auc_log:.4f}")
```

```python
fpr, tpr, _ = roc_curve(y_test, y_proba_log)
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f'Logistic ROC AUC = {roc_auc_log:.3f}', linewidth=2)
plt.plot([0,1],[0,1], linestyle='--', alpha=0.6)
plt.legend()
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC Curve')
plt.show()
```

# Random Forest: baseline then hyperparameter tuning

```python
# Start with a small baseline RandomForest, then run RandomizedSearchCV
# to find better hyperparameters. We use n_jobs=-1 and a modest cv for speed.
rf_base = RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE, n_jobs=-1)
rf_base.fit(X_train_raw, y_train)   # using raw features (RandomForest is scale-invariant)

y_pred_rf_base = rf_base.predict(X_test_raw)
print(f"RandomForest (baseline) Accuracy: {accuracy_score(y_test, y_pred_rf_base)*100:.2f}%")
```

```python
# Confusion matrix - baseline RF
cm_rf_base = confusion_matrix(y_test, y_pred_rf_base)
plt.figure(figsize=(6,5))
sns.heatmap(cm_rf_base, annot=True, fmt='d', cmap='YlGnBu', linewidths=1.5, linecolor='black')
plt.title("RandomForest Baseline — Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

### RandomizedSearchCV for RandomForest hyperparameters

```python
# Define parameter distribution and run RandomizedSearchCV. n_iter and cv are chosen to
# balance search thoroughness & speed. Increase n_iter/cv if you have more compute/time.
param_dist = {
    'n_estimators': [50, 100, 200, 300, 400, 500],
    'max_depth': [None, 5, 10, 15, 20],
    'max_features': randint(3, X_raw.shape[1]+1),
    'min_samples_leaf': randint(1, 5),
    'criterion': ['gini', 'entropy'],
    'bootstrap': [True, False]
}

rf = RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1)

search = RandomizedSearchCV(
    rf,
    param_distributions=param_dist,
    n_iter=40,         # change to e.g. 80 if you want a more thorough search
    cv=5,
    scoring='accuracy',
    random_state=RANDOM_STATE,
    n_jobs=-1,
    verbose=1
)

print("Starting RandomizedSearchCV for RandomForest (this may take a few minutes)...")
search.fit(X_train_raw, y_train)   # RF uses raw features

print("\nBest hyperparameters found:")
print(search.best_params_)
print(f"Best CV accuracy: {search.best_score_:.4f}")
```

### Train best RandomForest and evaluate

```python
# Use the best estimator returned by RandomizedSearchCV, evaluate on test set and present colorful plots.
best_rf = search.best_estimator_
best_rf.fit(X_train_raw, y_train)

y_pred_rf = best_rf.predict(X_test_raw)
rf_acc = accuracy_score(y_test, y_pred_rf)
print(f"Tuned RandomForest Test Accuracy: {rf_acc*100:.2f}%")
print("\nClassification Report (RandomForest):")
print(classification_report(y_test, y_pred_rf, digits=4))

# Confusion matrix - tuned RF
cm_rf = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(6,5))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='viridis', linewidths=1.5, linecolor='black')
plt.title("Tuned RandomForest — Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# ROC AUC for RF
if hasattr(best_rf, "predict_proba"):
    y_proba_rf = best_rf.predict_proba(X_test_raw)[:,1]
    roc_auc_rf = roc_auc_score(y_test, y_proba_rf)
    print(f"RandomForest ROC AUC: {roc_auc_rf:.4f}")

    fpr_rf, tpr_rf, _ = roc_curve(y_test, y_proba_rf)
    plt.figure(figsize=(6,5))
    plt.plot(fpr_rf, tpr_rf, label=f'RandomForest ROC AUC = {roc_auc_rf:.3f}', linewidth=2)
    plt.plot([0,1],[0,1], linestyle='--', alpha=0.6)
    plt.legend()
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('RandomForest ROC Curve')
    plt.show()
else:
    print("Best RF does not support predict_proba — can't compute ROC AUC")
```

### Feature importance 

```python
# Show the top features the RandomForest considers important. This helps interpretability.
importances = pd.Series(best_rf.feature_importances_, index=X_raw.columns).sort_values(ascending=False)
top_n = 15
plt.figure(figsize=(8,6))
sns.barplot(x=importances.iloc[:top_n], y=importances.iloc[:top_n].index, palette='rocket')
plt.title(f"Top {top_n} Feature Importances (RandomForest)")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.show()

print("\nTop 15 feature importances:")
display(importances.head(top_n))
```

### Persist (pickle) the tuned RandomForest

```python
# Save the tuned RandomForest model to disk. This can be loaded later using pickle.load.
model_filename = "cancer_rf.pkl"
with open(model_filename, "wb") as f:
    pickle.dump(best_rf, f)

print(f"Saved tuned RandomForest model to '{model_filename}'")

# (Optional) Save the scaler too if you plan to apply Logistic Regression later in production
scaler_filename = "scaler.pkl"
with open(scaler_filename, "wb") as f:
    pickle.dump(scaler, f)
print(f"Saved StandardScaler to '{scaler_filename}' (optional)")
```

### Quick example: load the pickle and predict on a sample

```python
# Demonstrates how to load the saved model and make predictions on new data.
with open(model_filename, "rb") as f:
    loaded_rf = pickle.load(f)

sample = X_test_raw.iloc[:5]   # first five test rows (raw features)
print("Sample rows (raw):")
display(sample)

print("Loaded RF predictions (0=benign,1=malignant):")
print(loaded_rf.predict(sample))
print("Actual labels:")
print(y_test.iloc[:5].values)
```
